```python
df_sorted = df.sort_values('yield', ascending=False)
df_inner_ad = df_sorted[df_sorted['AD'] > 0]
```

```python
df_origin = pd.read_csv('./data/structure_creation3/smiles.csv',index_col=0)
df_origin
```

```python
from rdkit import Chem
from rdkit.Chem import Draw

def draw(smiles):
    print(smiles)
    mol = Chem.MolFromSmiles(smiles)
    display(Draw.MolToImage(mol,size=(150, 150)))
```

```python
def save(smiles,filename):
    mol = Chem.MolFromSmiles(smiles)
    Chem.MolToMolFile(mol,f'{filename}.mol')
```

```python
for i,index in enumerate(df_inner_ad.index):
    smiles = df_origin.loc[index,:]['SMILES']
    print(index,'=',smiles)
    print('yield=', df_inner_ad.loc[index,'yield'])
    draw(smiles)
    save(smiles,f'./data/result3/structure/{index}')
    
    if (i == 20):
        break
    
```
